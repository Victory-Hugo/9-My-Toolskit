```python
import pandas as pd
```

```python
df = pd.read_csv('/mnt/d/迅雷下载/鲍曼组装/merged.csv',low_memory=False)
df
```

```python
# 由于 df["publication_date"] 符合标准ISO格式，可以直接使用方法
df["publication_date"] = pd.to_datetime(df["publication_date"])
```

```python
# 分拆为 年/月/日/时/分/秒
df["year"] = df["publication_date"].dt.year
df["month"] = df["publication_date"].dt.month
df["day"] = df["publication_date"].dt.day
df["hour"] = df["publication_date"].dt.hour
df["minute"] = df["publication_date"].dt.minute
df["second"] = df["publication_date"].dt.second
```

```python
df
```

```python
# 先对 geo_loc_name 做 split
split_cols = df["geo_loc_name"].str.split(":", n=1, expand=True)
# 生成新列
df["Country"] = split_cols[0].str.strip()
df["Location"] = split_cols[1].str.strip()
```

```python
df
```

```python
def parse_lat_lon(value):
    if pd.isna(value) or value == "NA":
        return (None, None)
    try:
        parts = value.split()
        lat, lat_dir, lon, lon_dir = parts
        lat, lon = float(lat), float(lon)

        # 根据方向修正正负号
        if lat_dir.upper() == "S":
            lat = -lat
        if lon_dir.upper() == "W":
            lon = -lon
        return (lat, lon)
    except Exception:
        return (None, None)
```

```python
# 应用函数
df[["Latitude", "Longitude"]] = df["lat_lon"].apply(lambda x: pd.Series(parse_lat_lon(x)))
```

```python
df.to_csv('/mnt/d/迅雷下载/鲍曼组装/merged_clean.csv',index=False)
```
