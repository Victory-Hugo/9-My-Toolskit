```python
import pandas as pd
```

```python
LOCAL_FILE = "/mnt/f/OneDrive/文档（科研）/脚本/Download/9-My-Toolskit/1-下载数据/script/2-ENA/conf/md5.txt"
CLOUD_FILE = "/mnt/f/OneDrive/文档（共享）/4_古代DNA/ERR_aDNA_MD5.txt"
```

```python
df_local = pd.read_csv(LOCAL_FILE, sep="\s+", header=None, names=["md5", "File_Path"])
df_local
# 将File_Path 列中的文件名提取出来，按照'/'分割
df_local['File_Name'] = df_local['File_Path'].str.split('/').str[-1]
# 将File_Name 中的基本名提取出来，去掉'.fastq.gz'后缀
df_local['File_Name'] = df_local['File_Name'].str.replace('.fastq.gz', '', regex=False)

df_local = df_local[['File_Name', 'md5']]
df_local
```

```python
df_cloud = pd.read_csv(CLOUD_FILE, sep="\t")
# 只保留需要的列
# 重命名列以匹配本地文件格式
df_cloud = df_cloud.rename(columns={"run_accession": "File_Name", "fastq_md5": "md5"})
#! 需要处理双端测序，双端测序情况下，File_Name包括_1和_2
#! 双端测序在File_Name列只有1行，但是fastq_ftp和md5有2列，由;分割，前一个是_1，后一个是_2
# 拆分 fastq_ftp 和 md5，先转成 DataFrame 的 list
df_cloud['fastq_ftp_new'] = df_cloud['fastq_ftp'].str.split(';')
df_cloud['md5_new'] = df_cloud['md5'].str.split(';')
df_cloud
```

```python
# 展开长数据（每个文件一行）
df_long = df_cloud.explode(['fastq_ftp_new', 'md5_new'], ignore_index=True)
df_long
```

```python
# 只保留你需要的列
df_long = df_long[['File_Name','fastq_ftp_new', 'md5_new']]
```

```python
df_long_trim = df_long.copy()
# 按照fastq_ftp_new拆分,按照/拆分，取最后一列
df_long_trim['File_Name'] = df_long['fastq_ftp_new'].str.split('/').str[-1]
# 去掉.fastq.gz后缀
df_long_trim['File_Name'] = df_long_trim['File_Name'].str.replace('.fastq.gz', '', regex=False)
df_long_trim = df_long_trim.loc[:,['File_Name', 'md5_new']]
df_cloud_new = df_long_trim
```

```python
# 去重
df_cloud_new = df_cloud_new.drop_duplicates(subset=['File_Name', 'md5_new'])
df_local = df_local.drop_duplicates(subset=['File_Name', 'md5'])
df_local['Type'] = 'Local'
df_cloud_new['Type'] = 'Cloud'
df_cloud_new,df_local
```

```python
# 1. 合并
df_merge = pd.merge(df_cloud_new, df_local, left_on='File_Name', right_on='File_Name', how='outer', suffixes=('_cloud', '_local'))

# 2. 标记情况
# 本地和云端都存在
mask_both = df_merge['md5_new'].notnull() & df_merge['md5'].notnull()
# MD5一致
mask_md5_equal = mask_both & (df_merge['md5_new'] == df_merge['md5'])
# MD5不一致
mask_md5_not_equal = mask_both & (df_merge['md5_new'] != df_merge['md5'])
# 仅云端有（本地没下载）
mask_only_cloud = df_merge['md5_new'].notnull() & df_merge['md5'].isnull()
# 仅本地有（一般不需处理，但可以输出）
mask_only_local = df_merge['md5_new'].isnull() & df_merge['md5'].notnull()

# 3. 筛选三类
df_md5_match = df_merge[mask_md5_equal]      # 下载成功
df_md5_mismatch = df_merge[mask_md5_not_equal]  # 有但损坏或错误
df_need_download = df_merge[mask_only_cloud]    # 需要补充下载
```

```python
df_md5_match_ID = df_md5_match.loc[:,['File_Name']]
df_md5_mismatch_ID = df_md5_mismatch.loc[:,['File_Name']]
df_need_download_ID = df_need_download.loc[:,['File_Name']]
```

```python
OUTDIR = "/mnt/f/OneDrive/文档（科研）/脚本/Download/9-My-Toolskit/1-下载数据/script/2-ENA/debug/"
```

```python
pd.Series(df_md5_match_ID['File_Name'].str.split('_').str[0].unique()).sort_values().to_csv(
    OUTDIR + "md5_完全匹配.txt", sep="\t", index=False, header=False
)
pd.Series(df_md5_mismatch_ID['File_Name'].str.split('_').str[0].unique()).sort_values().to_csv(
    OUTDIR + "md5_损坏.txt", sep="\t", index=False, header=False
)
pd.Series(df_need_download_ID['File_Name'].str.split('_').str[0].unique()).sort_values().to_csv(
    OUTDIR + "md5_缺失.txt", sep="\t", index=False, header=False
)
df1 = pd.Series(df_md5_mismatch_ID['File_Name'].str.split('_').str[0].unique())
df2 = pd.Series(df_need_download_ID['File_Name'].str.split('_').str[0].unique())
out = pd.concat([df1, df2], ignore_index=True).drop_duplicates().sort_values()
out.to_csv(OUTDIR + "md5_损坏或缺失.txt", sep="\t", index=False, header=False)
```
